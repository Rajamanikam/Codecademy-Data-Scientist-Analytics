# Exploring Student Data

Imagine that you work for a school district and have collected some data on local students and their parents. You’ve been tasked with answering some important questions:

1. How are students performing in their math classes?
2. What do students’ parents do for work?
3. How often are students absent from school?


```python
# Load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
students = pd.read_csv('students.csv')

# Print first few rows of data

print(students.head())

# Print summary statistics for all columns

print("\n*******************************")
print("SUMMARY STATISTICS:")
print("*******************************\n")

print(students.describe(include='all'))
```

```python
# Central tendency & Spread Statistics of math_grade column


# Calculate mean

mean_value = students['math_grade'].mean()

print(mean_value)


# Calculate median

median_value = students['math_grade'].median()

print(median_value)


# Calculate mode

mode_value = students['math_grade'].mode()[0]

print(mode_value)


# Calculate range

grade_range = students['math_grade'].max() - students['math_grade'].min()

print(grade_range)


# Calculate standard deviation

stand_dev = students['math_grade'].std()

print(stand_dev) #STD: 4.6


# Calculate MAD # mad() is depreciated in latest pandas versions

# Calculate the absolute deviations
absolute_deviation = (students['math_grade'] - mean_value).abs()

# Calculate the mean of the absolute deviations
mad_value = absolute_deviation.mean()

print(mad_value)
```

```python
# Create a histogram of math grades

sns.histplot(x='math_grade', data = students)

plt.show()
plt.clf()

# Create a box plot of math grades

sns.boxplot(x = 'math_grade', data = students)

plt.show()
plt.clf()
```

```python
# Calculate number of students with mothers in each job category

print(students['Mjob'].value_counts())

# Calculate proportion of students with mothers in each job category

print(students['Mjob'].value_counts(normalize = True))

# Create bar chart of Mjob

sns.countplot(x = 'Mjob', data = students)

plt.show()
plt.clf()

# Create pie chart of Mjob

students.Mjob.value_counts().plot.pie()

plt.show()
```

```python
# Calculate number of students with fathers in each job category

print(students['Fjob'].value_counts())

# Calculate proportion of students with mothers in each job category

print(students['Fjob'].value_counts(normalize = True))

# Create bar chart of Fjob

sns.countplot(x = 'Fjob', data = students)

plt.show()
plt.clf()

# Create pie chart of Mjob

students.Fjob.value_counts().plot.pie()

plt.show()
```

```python
# Calculate number of students in different address

print(students['address'].value_counts())

# Calculate proportion of students in different address

print(students['address'].value_counts(normalize = True))

# Create bar chart of address type

sns.countplot(x = 'address', data = students)

plt.show()
plt.clf()

# Create pie chart of Mjob

students.address.value_counts().plot.pie()

plt.show()
```

```python
# Central tendency & Spread Statistics of absences column


# Calculate mean

mean_value = students['absences'].mean()

print(mean_value)


# Calculate median

median_value = students['absences'].median()

print(median_value)


# Calculate mode

mode_value = students['absences'].mode()[0]

print(mode_value)


# Calculate range

absent_range = students['absences'].max() - students['absences'].min()

print(absent_range)


# Calculate standard deviation

stand_dev = students['absences'].std()

print(stand_dev) #STD: 8.0


# Calculate MAD # mad() is depreciated in latest pandas versions

# Calculate the absolute deviations
absolute_deviation = (students['absences'] - mean_value).abs()

# Calculate the mean of the absolute deviations
mad_value = absolute_deviation.mean()

print(mad_value)
```
